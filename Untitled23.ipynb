```python
!pip install yfinance pandas
import yfinance as yf
import pandas as pd

def download_amazon_data():
    try:
        # Initialisation du ticker pour Amazon
        amzn = yf.Ticker("AMZN")

        # RГ©cupГ©ration de l'historique complet depuis la crГ©ation
        print("TГ©lГ©chargement des donnГ©es historiques pour AMZN...")
        df = amzn.history(period="max")

        if df.empty:
            print("Aucune donnГ©e trouvГ©e.")
            return

        # Sauvegarde en CSV
        file_name = "amazon_stock_history.csv"
        df.to_csv(file_name)

        print(f"SuccГЁs ! {len(df)} lignes ont Г©tГ© sauvegardГ©es dans {file_name}")
        return df.head()

    except Exception as e:
        print(f"Une erreur est survenue : {e}")

# ExГ©cution du script
if __name__ == "__main__":
    download_amazon_data()
```

```python
!pip install numpy matplotlib scikit-learn tensorflow pandas

# Importation des bibliothГЁques
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input
from tensorflow.keras.callbacks import EarlyStopping

# 1. Chargement et prГ©paration des donnГ©es
df = pd.read_csv('amazon_stock_history.csv', index_col='Date', parse_dates=True)
data = df[['Close']].values

# Normalisation
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# ParamГЁtres de la sГ©quence
prediction_days = 60

x_train, y_train = [], []

# CrГ©ation des sГ©quences d'entraГ®nement (80% des donnГ©es)
train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[0:train_size, :]

for x in range(prediction_days, len(train_data)):
    x_train.append(train_data[x-prediction_days:x, 0])
    y_train.append(train_data[x, 0])

x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))

# 2. Construction du modГЁle LSTM optimisГ©
model = Sequential([
    Input(shape=(x_train.shape[1], 1)),
    LSTM(units=50, return_sequences=True),
    Dropout(0.2),
    LSTM(units=50, return_sequences=False),
    Dropout(0.2),
    Dense(units=25),
    Dense(units=1)
])

model.compile(optimizer='adam', loss='mean_squared_error')

# 3. EntraГ®nement pour une stabilisation maximale
# Augmentation Г  500 epochs et patience Г  25
early_stop = EarlyStopping(monitor='val_loss', patience=25, restore_best_weights=True)

model.fit(
    x_train, y_train,
    epochs=500,
    batch_size=32,
    validation_split=0.1,
    callbacks=[early_stop]
)

# 4. PrГ©diction sur les donnГ©es de test
test_data = scaled_data[train_size - prediction_days:, :]
x_test = []
y_test = data[train_size:, :]

for x in range(prediction_days, len(test_data)):
    x_test.append(test_data[x-prediction_days:x, 0])

x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)

# 5. Visualisation
plt.figure(figsize=(12,6))
plt.plot(y_test, color='black', label="Prix RГ©el")
plt.plot(predictions, color='green', label="PrГ©diction LSTM")
plt.title("PrГ©diction Amazon - Stabilisation Maximale (500 Epochs Max)")
plt.xlabel("Temps (Index Test)")
plt.ylabel("Prix de l'action")
plt.legend()
plt.show()
```

```python
import pickle

# Sauvegarde du modГЁle
model.save('amazon_lstm_model.keras')

# Sauvegarde du scaler (indispensable pour les futures prГ©dictions)
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

print("ModГЁle et scaler sauvegardГ©s avec succГЁs.")
```
